```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread

imagem_colorida = imread('Lenna_(test_image).png')

imagem_cinza = 0.299 * imagem_colorida[:, :, 0] + 0.587 * imagem_colorida[:, :, 1] + 0.114 * imagem_colorida[:, :, 2]

limiar = 0.5

if imagem_cinza.max() > 1:
    imagem_cinza = imagem_cinza / 255.0

imagem_binaria = (imagem_cinza > limiar).astype(np.uint8)

plt.imshow(imagem_binaria, cmap='gray')
plt.axis('off')
plt.show()
```
